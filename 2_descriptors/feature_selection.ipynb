```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
df = pd.read_csv('/home/student/Desktop/enam/data_preprocessing/2_all_descriptors.csv', low_memory=False)

print(df.info())
```

```python
print(df)
```

```python
df_cleaned = df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1', 'mol_id'], axis=1)
df_cleaned.drop_duplicates(subset=['smiles'], inplace=True)
df_cleaned.reset_index(drop=True, inplace=True)
print(df_cleaned)
```

```python
df_cleaned.drop(columns=['homo', 'lumo'], inplace=True)
df_cleaned.info()
```

```python
# Находим столбцы с типом данных bool
bool_columns = df_cleaned.select_dtypes(include=['bool']).columns

# Преобразуем значения в числовой формат
df_cleaned[bool_columns] = df_cleaned[bool_columns].astype(int)

# Кодирование SMILES
label_encoder = LabelEncoder()
df_cleaned['smiles_encoded'] = label_encoder.fit_transform(df_cleaned['smiles'])
df_cleaned.drop(columns=['smiles'], axis=1, inplace=True)

# Получение списка колонок с типами данных, отличными от float и int
non_numeric_columns = df_cleaned.select_dtypes(exclude=['float', 'int']).columns.tolist()

print("Следующие колонки имеют тип данных, отличный от float или int:")
print(non_numeric_columns)
```

```python
print(df_cleaned[non_numeric_columns].info())
print(df_cleaned[non_numeric_columns])
```

```python
# Удаление некорректно полученных данных
deleted_columns = []
for column in non_numeric_columns:
    if any(pd.to_numeric(df_cleaned[column], errors='coerce').isnull()):
        deleted_columns.append(column)
        df_cleaned = df_cleaned.drop(column, axis=1)

print(df_cleaned.info())
print(deleted_columns)
```

```python
# Проверяем наличие NaN значений в датасете
nan_rows = df_cleaned[df_cleaned.isnull().any(axis=1)]
print("Количество строк с NaN значениями:")
print(len(nan_rows))
```

```python
# Фильтрация по коэффициенту корреляции
"""
Мы хотим убрать дескрипторы, коррелирующие между собой, но их корреляция с целевой характеристикой важна
Поэтому мы уберем из датасета gap на время фильтрации
"""
df_gap = pd.DataFrame()
df_gap['gap'] = df_cleaned['gap']
df_gap['smiles_encoded'] = df_cleaned['smiles_encoded']
df_cleaned.drop(columns=['gap'], inplace=True)

# Вычисление матрицы корреляции
corr_matrix = df_cleaned.corr()

# Выбор дескрипторов с корреляцией > 0.75 или < -0.75
high_corr_descriptors = set()
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > 0.75:
            high_corr_descriptors.add(corr_matrix.columns[i])

# Удаление выбранных дескрипторов
data_filtered = df_cleaned.drop(high_corr_descriptors, axis=1)

print("Удалены следующие дескрипторы с высокой корреляцией:")
print(high_corr_descriptors)
```

```python
merged_df = pd.merge(data_filtered, df_gap[['smiles_encoded', 'gap']], on='smiles_encoded', how='left')
```

```python
# Определяем столбцы с NaN значениями
nan_columns = merged_df.columns[merged_df.isnull().any()]
print("Столбцы с NaN значениями:")
print(nan_columns)
```

```python
# Удаление столбцов с нулевой вариацией
merged_df = merged_df.loc[:, merged_df.std() > 0]

merged_df.info()
```
